```python
def add(a,b):
    print(a+b)

def sub(a,b):
    print(a-b)
    
def mul(a,b):
    print(a*b)

def div(a,b):
    print(a/b)
    
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")

    if o == '+':
        add(a,b)
    elif o == '-':
        sub(a,b)
    elif o == '*':
        mul(a,b)
    elif o == '/':
        div(a,b)
    else:
        print("Invalid operator")
 

cal()
```

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b
    
def mul(a,b):
    return a*b

def div(a,b):
    return a/b
    
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")

    if o == '+':
        print(add(a,b))
    elif o == '-':
        print(sub(a,b))
    elif o == '*':
        print(mul(a,b))
    elif o == '/':
        print(div(a,b))
    else:
        print("Invalid operator")
 
cal()
```

```python
def cal(l,b,h):
    a = l*b
    v = a*h
    return a,v
    
l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))
area,volume = cal(l,b,h)

print("The area is = ",area)
print("The volume = ",volume)
```
